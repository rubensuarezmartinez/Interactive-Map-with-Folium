```python
# TASK 3: Calculate the distances between a launch site to its proximities
proximities = [
    {'Location': 'Proximity 1', 'Latitude': 29.1234, 'Longitude': -95.6789},
    {'Location': 'Proximity 2', 'Latitude': 29.5678, 'Longitude': -95.4321},
    # Add more proximities as needed
]

from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in kilometers
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    # Calculate differences
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    # Haversine formula
    a = sin(dlat/2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

# Dictionary to store distances between launch sites and their proximities
launch_site_distances = {}

# Iterate through each launch site
for index, site in launch_sites_df.iterrows():
    site_distances = {}
    # Iterate through proximities of the current launch site
    for proximity in proximities:
        # Calculate distance between launch site and proximity
        distance = calculate_distance(site['Latitude'], site['Longitude'], proximity['Latitude'], proximity['Longitude'])
        # Store distance in site_distances dictionary
        site_distances[proximity['Location']] = distance
    # Store site_distances in launch_site_distances dictionary
    launch_site_distances[site['LaunchSite']] = site_distances

# Print launch_site_distances dictionary
print(launch_site_distances)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
